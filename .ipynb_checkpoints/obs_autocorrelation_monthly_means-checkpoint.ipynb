```python
import numpy as np
import pandas as pd
import xarray as  xr
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import math
```

```python
obs_name = 'NSIDC_0079'
obs_type = 'sipn_nc_yearly_agg'
no_day_change = 5
obs_filename = '/home/disk/sipn/mcmcgraw/data/VRILE/{model_name}_{model_type}_SIE_d_SIE_{d_days}day_change_lead_time_ALL_REGIONS_ALL_ENS.csv'.format(model_name=obs_name,
                       model_type=obs_type,d_days=no_day_change)
SIE_obs = pd.read_csv(obs_filename)
SIE_obs = SIE_obs.dropna()
```

```python
regions = SIE_obs['region'].unique().tolist()
i_reg = 0
SIE_obs_trim = SIE_obs.where(SIE_obs['region']==regions[i_reg])
SIE_obs_trim = SIE_obs_trim.dropna()
SIE_obs_trim.head()
```

```python
yr = np.arange(1989,2019)
SIE_obs_monthly_means = pd.DataFrame(columns=['Year','Month','mean SIE','region','ensemble'])
for yr_sel in yr:
    df_yr = pd.DataFrame(columns=['Year','Month','mean SIE','region','ensemble'])
    SIE_yr_trim = SIE_obs_trim.where(pd.to_datetime(SIE_obs_trim['V (valid date)']).dt.year == yr_sel)
    #pd.to_datetime(SIE_obs_trim['V (valid date)']).dt.year
    mon_means = SIE_yr_trim.groupby(['V_mon (valid date month)']).mean()
    df_yr['mean SIE'] = mon_means['SIE']
    df_yr['Year'] = np.tile(yr_sel,len(mon_means))
    df_yr['Month'] = np.arange(1,13)
    df_yr['region'] = np.tile(regions[i_reg],len(mon_means))
    df_yr['ensemble'] = np.tile('obs',len(mon_means))
    SIE_obs_monthly_means = SIE_obs_monthly_means.append(df_yr)
```

```python
SIE_obs_monthly_means.iloc[0::12,:]
```

```python
SIE_seas_cyc = SIE_obs_monthly_means.groupby(['Month']).transform('mean')
SIE_seas_anoms = SIE_obs_monthly_means['mean SIE'] - SIE_seas_cyc['mean SIE']
```

```python
fart = sm.tsa.acf(SIE_seas_anoms,nlags=36)
plt.plot(fart)
```
