Load obs

```python
obs_name = 'NSIDC_0079'
obs_type = 'sipn_nc_yearly_agg'
obs_filename = '/home/disk/sipn/mcmcgraw/data/VRILE/{model_name}_{model_type}_SIE_d_SIE_{d_days}day_change_lead_time_ALL_REGIONS_ALL_ENS.csv'.format(model_name=obs_name,
                       model_type=obs_type,d_days=no_day_change)
SIE_obs = pd.read_csv(obs_filename)
SIE_obs = SIE_obs.dropna()
```

Choose a region

```python
regions = d_SIE_all['region'].unique().tolist()
lead_days = d_SIE_all['lead time (days)'].unique().tolist()
```

Group model data by lead day and choose a maximum lead. This code will include all days up to and included the maximum lead day.

```python
lead_day_choose = 28
lead_day_sel = np.arange(0,lead_day_choose+1)
lead_day_max = lead_day_sel[-1]
f_SIE_lead_sel = d_SIE_all['lead time (days)'].isin(lead_day_sel)
SIE_lead_sel = d_SIE_all[f_SIE_lead_sel]
SIE_model_dates = pd.to_datetime(SIE_lead_sel['V (valid date)'])
mon_choose = [6,7,8,9]
MON_STR = 'JJAS'
f_seas_sel = SIE_model_dates.dt.month.isin(mon_choose)
SIE_seas_sel = SIE_lead_sel[f_seas_sel]
SIE_model_dates = SIE_model_dates[f_seas_sel]
print('max lead day is ',lead_day_max)
#pd.to_datetime(SIE_lead_sel['V (valid date)']).dt.month
print('season is ',MON_STR)
```

Keep only observations that match dates associated with model lead times.  

```python
SIE_obs_dates = pd.to_datetime(SIE_obs['V (valid date)'])
f_dates = SIE_obs_dates.isin(SIE_model_dates)
SIE_obs_trim = SIE_obs.loc[f_dates]
SIE_obs_trim.head()
```

Trim regions

```python
#groupby region
SIE_mod_reg_groups = SIE_seas_sel.groupby(['region'])
SIE_obs_reg_groups = SIE_obs_trim.groupby(['region'])
dates_list = SIE_seas_sel['V (valid date)'].unique().tolist()
dates_list_obs = pd.to_datetime(SIE_obs_trim['V (valid date)'])
##Start by looping through regions
#for ireg in np.arange(0,len(regions)):
ireg = 2
region_sel = regions[ireg]
print('running {region}'.format(region=region_sel))
SIE_model_sel = SIE_mod_reg_groups.get_group(region_sel)
SIE_obs_sel = SIE_obs_reg_groups.get_group(region_sel)
SIE_obs_sel['V (valid date)'] = pd.to_datetime(SIE_obs_sel['V (valid date)'])
```

Calculate 5th percentiles of d_SIE for model and for obs

```python
THRESH = 0.05
pct_obs = SIE_obs_sel['SIE'].quantile(THRESH)
pct_mod = SIE_model_sel['SIE'].quantile(THRESH)
print('observed threshold: ',pct_obs)
print('model threshold: ',pct_mod)
#fart = pd.Timestamp(SIE_obs_sel['V (valid date)'])
#SIE_obs_sel.iloc[2:100,:]
```

Now, calculate Brier score as a function of lead time.  The Brier score means that $SIE_x$ will be 1 when we observe an extreme event (d_SIE less than threshhold), or 0 otherwise.  Brier score: <br>
$BS = [\sum_{n=1}^{e} SIE_e(I,V) - SIE_{obs}(V)]^2$ <br>
where $n$ is the ensemble (1 to $e$), $I$ and $V$ refer to the initialization date (represented by lead time) and the valid date. Thus for each valid date, at each lead time, we calculate the $BS$ (either 0 or 1) for each ensemble member, and then we sum them up and normalize by $e$, the number of ensembles.  We then square the difference between the modeled probabilistic BS (between 0 and 1) and the observed BS (either 0 or 1).  <br>
So, if we observe an extreme event and 19/20 ensemble members also observe it, our BS for the model will be very close to 1, the difference  bewteen the two will be small, and we'll have high skill.  

```python
SIE_group_valid = SIE_model_sel.groupby(['V (valid date)'])
SIE_obs_group_valid = SIE_obs_sel.groupby(['V (valid date)'])
for i_date in dates_list:
#i_date = dates_list[0]
#i_date_obs = dates_list_obs[10]
    SIE_model_i_date = SIE_group_valid.get_group(i_date).dropna()
    SIE_obs_i_date = SIE_obs_group_valid.get_group(i_date).dropna()
    #Calculate observed skill score--False = 0,  True = 1
    obs_score = (SIE_obs_i_date['SIE'].values <= pct_obs).sum()
    #Group model data by lead time
    SIE_model_i_date_groups = SIE_model_i_date.groupby(['lead time (days)'])
    #Calculate model skill score: False = 0, True = 1, and we use mean to divide by the total count
    model_score = SIE_model_i_date_groups.agg({'SIE':lambda x: np.mean(x <= pct_mod)})
    #Calculate Brier score--(model - obs)^2
    brier_score = (model_score - obs_score).pow(2)
    #Create a dataframe where we index by both valid date and lead time
    idx = pd.MultiIndex.from_product([[i_date],
                                      np.sort(SIE_model_i_date['lead time (days)'].unique())],
                                     names=['V (valid date)', 'lead time (days)'])
    col = ['brier score']
    df = pd.DataFrame('-', idx, col)
    df['brier score'] = brier_score['SIE'].values
    if i_date == dates_list[0]:
        SIE_brier_score = df.copy()
    else:
        SIE_brier_score = SIE_brier_score.append(df)
```

```python
f_lead = np.isin(lead_days,lead_day_sel)
lead_day_vec = np.array(lead_days)[f_lead]
mean_brier_d_SIE = np.empty([])
for x_lead in lead_day_vec:
    x_brier = SIE_brier_score.xs(x_lead,level='lead time (days)',drop_level=False).mean()
    mean_brier_d_SIE = np.append(mean_brier_d_SIE,x_brier['brier score'])
#    mean_ensemble_spread_d_SIE = np.append(mean_ensemble_spread_d_SIE,x_RMSE['d_SIE spread'])
```
